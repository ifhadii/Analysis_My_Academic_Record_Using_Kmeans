```python
import tabula
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def convert_pdf_to_csv(pdf_path, output_csv):
    """
    Convert a PDF file to CSV.

    Parameters:
        pdf_path (str): Path to the PDF file.
        output_csv (str): Path to the output CSV file.

    Returns:
        None
    """
    try:
        # Read the PDF file and extract tables
        tables = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True)

        # Concatenate all tables into a single DataFrame
        combined_df = pd.concat(tables, ignore_index=True)

        # Write the combined DataFrame to a CSV file
        combined_df.to_csv(output_csv, index=False)
        
        print("Conversion completed successfully.")
    except Exception as e:
        print(f"Error: {e}")

# Example usage:
pdf_file = "202002381_19793.pdf"
output_csv = "output.csv"
convert_pdf_to_csv(pdf_file, output_csv)
```

```python
with open("output.csv", "r") as data:
    my_data1 = data.read()
    my_data2 = my_data1.split(',')
    my_data3 = ' '.join(my_data2)
    my_data4 = my_data3.split(' ')
print(my_data4)
```

```python
# List for grades 'A+'
ap = []

# List for grades 'A'
a = []

# List for grades 'B+'
bp = []

# List for grades 'B'
b = []

# List for grades 'C+'
cp = []

# List for grades 'C'
c = []

# List for grades 'D+'
dp = []

# List for grades 'D'
d = []

# List for grades 'F'
f = []
```

```python
# Orgnize for each catgorie 
for i in my_data4 :
    if i == "A+":
        ap.append(i)
    elif i == "A":
        a.append(i)
    elif i == "B+":
        bp.append(i)
    elif i == "B":
        b.append(i)
    elif i == "C+":
        cp.append(i)
    elif i == "C":
        c.append(i)
    elif i == "D+":
        dp.append(i)
    elif i == "D":
        d.append(i)
    elif i == "F":
        f.append(i)
```

```python
# Append Each Grade
print("Grade - A+:", ap)
print("Grade - A:", a)
print("Grade - B+:", bp)
print("Grade - B:", b)
print("Grade - C+:", cp)
print("Grade - C:", c)
print("Grade - D+:", dp)
print("Grade - D:", d)
print("Grade - F:", f)
```

```python
# Calculate Each Grade
print("Number Of A+ You Get:", len(ap))
print("Number Of A You Get:", len(a))
print("Number Of B+ You Get:", len(bp))
print("Number Of B You Get:", len(b))
print("Number Of C+ You Get:", len(cp))
print("Number Of C You Get:", len(c))
print("Number Of D+ You Get:", len(dp))
print("Number Of D You Get:", len(d))
print("Number Of F You Get:", len(f))
```

```python
# Combine all grade lists into a single list
all_grades = ap + a + bp + b + cp + c + dp + d + f

# Create a countplot
sns.countplot(x=all_grades)
plt.title('Distribution of Grades')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```

```python
# Combine all grade lists into a single list
all_grades = ap + a + bp + b + cp + c + dp + d + f

# Count the occurrences of each grade
grade_counts = {}
for grade in all_grades:
    if grade in grade_counts:
        grade_counts[grade] += 1
    else:
        grade_counts[grade] = 1

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(grade_counts.values(), labels=grade_counts.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Grades')
plt.show()
```

```python
# Create a swarm plot
sns.swarmplot(x=all_grades, size=5)
plt.title('Distribution of Grades')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```
